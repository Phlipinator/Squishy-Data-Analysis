```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None)   # show all columns
pd.set_option("display.max_rows", None)      # show all rows if needed
```

```python
# CLEANUP DATA

# Read file, skipping the first two metadata rows
df = pd.read_csv("SquishyData.csv", skiprows=[1,2])

# Drop the first 17 columns with unneeded data
df = df.iloc[:, 17:]

# Drop row 12, because of missing data and reset the indexing
df = df.drop(index=12)
df = df.reset_index(drop=True)

# Add a PID column
df.insert(0, "PID", range(1, len(df) + 1))
```

```python
geschlecht_map = {
    1: "Weiblich",
    2: "Männlich",
    3: "Divers",
    4: "Keine Angabe"
}

bildung_map = {
    1: "Kein Abschluss",
    2: "Hauptschulabschluss",
    3: "Realschulabschluss (Mittlere Reife)",
    4: "Fachhochschulreife (Abitur)",
    5: "Allgemeine Hochschulreife (Abitur)",
    6: "Berufsausbildung",
    7: "Bachelor",
    8: "Master",
    9: "Anderes"
}

techAff_KINutz_map = {
    1: "Sehr häufig",
    2: "Häufig",
    3: "Gelegentlich",
    4: "Selten",
    5: "Nie"
}

df["Geschlecht"] = df["Geschlecht"].replace(geschlecht_map)
df["Bildungsgrad"] = df["Bildungsgrad"].replace(bildung_map)
df["Technische Affinität"] = df["Technische Affinität"].replace(techAff_KINutz_map)
df["KI Nutzerverhalten"] = df["KI Nutzerverhalten"].replace(techAff_KINutz_map)

df
```
